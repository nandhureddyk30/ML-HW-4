```
# -------------------------------
# Q1: NLP Preprocessing Pipeline
# -------------------------------

import nltk
from nltk.corpus import stopwords, wordnet
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag

# Download required resources
nltk.download('punkt_tab')
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')

# Sentence input
sentence = "John enjoys playing football while Mary loves reading books in the library."

# Step 1: Break sentence into tokens
words = word_tokenize(sentence)

# Step 2: Remove stopwords and punctuation
stop_words = set(stopwords.words('english'))
clean_words = []

for w in words:
    if w.lower() not in stop_words and w.isalpha():
        clean_words.append(w)

# Step 3: POS tagging
tagged_words = pos_tag(clean_words)

# Convert POS to WordNet format
def convert_tag(tag):
    mapping = {
        'J': wordnet.ADJ,
        'V': wordnet.VERB,
        'N': wordnet.NOUN,
        'R': wordnet.ADV
    }
    return mapping.get(tag[0], wordnet.NOUN)

# Step 4: Lemmatization
lemmatizer = WordNetLemmatizer()
processed_words = []

for word, tag in tagged_words:
    lemma = lemmatizer.lemmatize(word, convert_tag(tag))

    # Step 5: Keep only nouns and verbs
    if tag.startswith('N') or tag.startswith('V'):
        processed_words.append(lemma)

# Final Output
print("Processed Words:", processed_words)
```

```
# -------------------------------
# Q2: NER + Pronoun Check
# -------------------------------

import spacy

# Load model
nlp = spacy.load("en_core_web_sm")

# Input text
data = "Chris met Alex at Apple headquarters in California. He told him about the new iPhone launch."

# Process text
doc = nlp(data)

# Extract entities into dictionary format
entities = {}

for ent in doc.ents:
    entities[ent.text] = ent.label_

print("Detected Entities:")
for name, label in entities.items():
    print(f"{name} -> {label}")

# Pronoun ambiguity detection (alternative method)
ambiguous_flag = False

for token in doc:
    if token.pos_ == "PRON":
        ambiguous_flag = True

if ambiguous_flag:
    print("\nAmbiguity Alert: Pronouns detected in sentence.")
```
